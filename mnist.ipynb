```typescript
import { mnist } from './src/nn/datasets.ts'
import { document } from 'jsr:@ry/jupyter-helper'
import { cell, plot } from 'npm:@observablehq/plot'

const [X_train, Y_train] = await mnist()

const index = 9944
const label = Y_train.get(index).tolist()
const data =  X_train.get(index).tolist<[number[][]]>()[0].flatMap((row, y) => row.map((value, x) => ({ x, y, value })))
plot({
  title: `Label: ${label}`,
  marks: [cell(data, { x: 'x', y: 'y', fill: 'value' })],
  color: { scheme: 'greys' },
  document,
})
```
